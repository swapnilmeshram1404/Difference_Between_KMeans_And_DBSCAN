### Lets create a function so that we dont have to repeat the steps because we are going to compare kmeans and dbscan

```python
def fun_cluster(model , data):
    
    labels = model.fit_predict(data)
    
    sns.scatterplot(data = data , x = "X1" , y = "X2" , 
                    hue = labels , palette= "Set1")
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
```

```python
fun_cluster(model , data_blobs)
```

```python
model = KMeans(n_clusters=2)
```

```python
fun_cluster(model , data_moons)
```

```python
fun_cluster(model , data_circles)
```

### Conclusion :

So we can see that the KMeans is using distance and hence not able to divide data into clusters properly for moons and circles dataset.

```python
from sklearn.cluster import DBSCAN
```

```python
model = DBSCAN()
```

```python
fun_cluster(model , data_blobs)
```

```python
model = DBSCAN(eps= 0.15)

fun_cluster(model , data_moons)
```

```python
fun_cluster(model , data_circles)
```
